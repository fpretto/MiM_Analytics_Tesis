## Preprocessing

```python
df = df[df['player_minutes']>=270].copy()
```

### P90 Stats

```python
df.head()
```

```python
# Attacking
df['offsides_p90'] = df['offsides']/df['player_minutes']*90
df['shots_p90'] = df['shots_total']/df['player_minutes']*90
df['shooting_accuracy'] = np.where(df['shots_total']==0, 0, df['shots_on_goal']/df['shots_total'])
df['goals_p90'] = df['goals_total']/df['player_minutes']*90
df['np_goals_p90'] = (df['goals_total']-df['penalty_scored'])/df['player_minutes']*90
df['goal_conversion'] = df['goals_total']/df['shots_total']
df['goal_conversion_nonpenalty'] = (df['goals_total']-df['penalty_scored'])/df['shots_total']
df['assists_p90'] = df['goals_assists']/df['player_minutes']*90
df['penalty_won_p90'] = df['penalty_won']/df['player_minutes']*90
df['penalty_missed_p90'] = df['penalty_missed']/df['player_minutes']*90
# Build-up
df['passes_p90'] = df['passes_total']/df['player_minutes']*90
df['passing_accuracy'] = np.where(df['passes_total']==0, 0, df['passes_completed']/df['passes_total'])
df['through_balls_p90'] = df['passes_key']/df['player_minutes']*90
df['fouls_drawn_p90'] = df['fouls_drawn']/df['player_minutes']*90
# Defense
df['tackles_p90'] = df['tackles_total']/df['player_minutes']*90
df['blocks_p90'] = df['tackles_blocks']/df['player_minutes']*90
df['interceptions_p90'] = df['tackles_interceptions']/df['player_minutes']*90
df['saves_p90'] = df['goals_saves']/df['player_minutes']*90
df['penalty_saves_p90'] = df['penalty_saved']/df['player_minutes']*90
df['penalty_committed_p90'] = df['penalty_committed']/df['player_minutes']*90
df['goals_conceded_p90'] = df['goals_conceded']/df['player_minutes']*90
df['fouls_committed_p90'] = df['fouls_committed']/df['player_minutes']*90
# One vs one
df['duels_p90'] = df['duels_total']/df['player_minutes']*90
df['duels_success_ratio'] = np.where(df['duels_total']==0, 0, df['duels_won']/df['duels_total'])
df['dribbles_p90'] = df['dribbles_attemps']/df['player_minutes']*90
df['dribbles_success_ratio'] = np.where(df['dribbles_attemps']==0, 0, df['dribbles_success']/df['dribbles_attemps'])
df['dribbles_past_ratio'] = np.where(df['dribbles_attemps']==0, 0, df['dribbles_past']/df['dribbles_attemps'])
# General
df['cards_yellow_p90'] = df['cards_yellow']/df['player_minutes']*90
df['cards_red_p90'] = df['cards_red']/df['player_minutes']*90
```

## Dash App

```python
cols=['player_name', 'player_preferred_position', 'player_preferred_number', 'team_name', 'league_name', 'country_name',
       'league_season', 'player_minutes', 'wavg_player_rating', 'goals_total', 'goals_assists']
```

```python
import dash
from dash.dependencies import Input, Output
import dash_table
import pandas as pd

df_dash = df[cols].rename(columns=dict_display_names).copy()

app = dash.Dash(__name__)

PAGE_SIZE = 75

app.layout = html.Div([
#     dcc.Slider(
#         id='goals-scored-slider',
#         min=df['goals_total'].min(),
#         max=df['goals_total'].max(),
#         value=df['goals_total'].min(),
#         marks={str(year): str(year) for year in df['goals_total'].unique()},
#         step=None
#     ),
    dash_table.DataTable(
            id='table-all-players',
            columns=[{"name": i, "id": i} for i in sorted(df_dash.columns)],
            data=df_dash.to_dict('records'),
            # Back-end Pagination
            page_current=0,
            page_size=PAGE_SIZE,
            page_action='custom',
            # Multi-sorting
            sort_action='custom',
            sort_mode='multi',
            sort_by=[],
            # Styling
            css=[{'selector': 'table-all-players', 'rule': 'table-layout: fixed'}],
            style_cell={
                'width': '{}%'.format(len(df_dash.columns)),
                'textOverflow': 'ellipsis',
                'overflow': 'hidden',
                'textAlign': 'center'
            },
            style_header={
                'backgroundColor': 'MidnightBlue', # LightSkyBlue
                'fontWeight': 'bold',
                'color': 'white'
            },
            style_as_list_view=True
        )
])

@app.callback(
    Output('table-all-players', "data"),
    Input('table-all-players', "page_current"),
    Input('table-all-players', "page_size"),
    Input('table-all-players', "sort_by"))
    #Input('goals-scored-slider', 'value'))

def update_table(page_current, page_size, sort_by):
    print(sort_by)
    if len(sort_by):
        dff = df_dash.sort_values(
            [col['column_id'] for col in sort_by],
            ascending=[col['direction'] == 'asc' for col in sort_by], inplace=False)
    else:
        # No sort is applied
        dff = df[cols]

    return dff.iloc[page_current*page_size:(page_current+1)*page_size].to_dict('records')

# def update_figure(slicer_value):
#     filtered_df = df[df['goals_total'] <= slicer_value][cols]

#     return filtered_df
    
if __name__ == '__main__':
    app.run_server(debug=False)
```
