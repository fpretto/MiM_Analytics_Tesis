### Countries

```python
# API Request
url = "https://v3.football.api-sports.io/countries"
response = requests.request("GET", url, headers=headers, data=payload)
```

```python
# Create DF
df_countries = pd.json_normalize(json.loads(response.text)["response"])
```

```python
# Enrichment: Add continent
lkp_countries_enrich = pd.read_csv(path+"INTERNET_country_and_continent_codes_list.csv", sep=",")
cols = ["Two_Letter_Country_Code", "Three_Letter_Country_Code", "Country_Number", "Continent_Code", "Continent_Name"]
df_countries = df_countries.merge(lkp_countries_enrich[cols], left_on="code", right_on="Two_Letter_Country_Code", how="left")
del df_countries["Two_Letter_Country_Code"]
df_countries.columns = ['country_name', 'country_code_2', 'country_flag', 'country_code_3', 'country_code_un', 'continent_code', 'continent_name']
```

```python
# NA Replacement
dict_fixes = {'XK': {'country_code_3': 'XKX', 'country_code_un': '383', 
                     'continent_code': 'EU', 'continent_name': 'Europe'},
              'NA': {'country_code_3': 'NAM', 'country_code_un': '516', 
                     'continent_code': 'AF', 'continent_name': 'Africa'}}

for country in dict_fixes.keys():
    filters = (df_countries['country_code_2']==country)
    df_countries.loc[filters, 'country_code_3'] = dict_fixes[country]['country_code_3']
    df_countries.loc[filters, 'country_code_un'] = dict_fixes[country]['country_code_un']
    df_countries.loc[filters, 'continent_code'] = dict_fixes[country]['continent_code']
    df_countries.loc[filters, 'continent_name'] = dict_fixes[country]['continent_name']
    
df_countries.loc[df_countries['continent_name']=='North America', 'continent_code'] = 'NA'
```

```python
# Export
cols = ['country_name', 'country_code_2', 'country_code_3', 'country_code_un', 'country_flag', 'continent_code', 'continent_name']
df_countries[cols].to_csv(path+"LKP_Countries.csv", encoding="utf-8", sep=",", index=False)
```

```python
def getAPIFootballData(endpoint, query=None):
    # since we have multiple API's (Pollution and Economy Data),
    # so we can get apt API link by passing in its name in function argument.

    if query == None:
        api_url = 'https://' + api_football['host'] + api_football['endpoints'][endpoint]
    else:
        api_url = 'https://' + api_football['host'] + api_football['endpoints'][endpoint] + query

    payload = {}
    headers = {'x-rapidapi-key': api_football['key'],
               'x-rapidapi-host': api_football['host']
               }

    response = requests.request("GET", api_url, headers=headers, data=payload).json()

    if len(response['errors']) > 1:
        raise Exception(f'El request {api_url} ha generado los siguientes errores: {response["errors"]}')
    else:
        return response
```

```python
# Home team statistics
home_team = pd.json_normalize(response["response"][0]["team"], sep="_")
home_team['fixture_id'] = fixture_id
home_team['fl_home'] = True
home_stats = pd.json_normalize(response["response"][0]["statistics"], sep="_").set_index(
    'type').transpose().reset_index(drop=True)
df_home = pd.concat([home_team[['fixture_id', 'fl_home', 'id', 'name']], home_stats.fillna(0)], axis=1)

# Away team statistics
away_team = pd.json_normalize(response["response"][1]["team"], sep="_")
away_team['fixture_id'] = fixture_id
away_team['fl_home'] = False
away_stats = pd.json_normalize(response["response"][1]["statistics"], sep="_").set_index(
    'type').transpose().reset_index(drop=True)
df_away = pd.concat([away_team[['fixture_id', 'fl_home', 'id', 'name']], away_stats.fillna(0)], axis=1)

# Match Statistics
df_match_stats = pd.concat([df_home, df_away])
df_match_stats.columns = [col.lower().replace(' ', '_') for col in df_match_stats.columns]

df_match_stats['ball_possession'] = df_match_stats['ball_possession'].apply(lambda x: float(str(x)[0:2]) / 100 if len(str(x)) > 2 else float(str(x)[0:1]) / 100)
df_match_stats['passes_pct'] = df_match_stats['passes_%'].apply(lambda x: float(str(x)[0:2]) / 100 if len(str(x)) > 2 else float(str(x)[0:1]) / 100)
del df_match_stats['passes_%']

cols_to_keep = ['fixture_id', 'id', 'name', 'fl_home', 'shots_on_goal', 'shots_off_goal',
                'total_shots', 'blocked_shots', 'shots_insidebox', 'shots_outsidebox', 'fouls', 'corner_kicks',
                'offsides', 'ball_possession', 'yellow_cards', 'red_cards', 'goalkeeper_saves', 'total_passes',
                'passes_accurate', 'passes_pct']
```

```python
# Home team statistics
home_team = pd.json_normalize(response["response"][0]["team"], sep="_")
home_team['match_id'] = fixture_id
home_team['home_away'] = 'Home'
home_stats = pd.json_normalize(response["response"][0]["statistics"], sep="_").set_index('type').transpose().reset_index(drop=True)
df_home = pd.concat([home_team[['match_id', 'home_away', 'id', 'name']], home_stats], axis=1)

# Away team statistics
away_team = pd.json_normalize(response["response"][1]["team"], sep="_")
away_team['match_id'] = fixture_id
away_team['home_away'] = 'Away'
away_stats = pd.json_normalize(response["response"][1]["statistics"], sep="_").set_index('type').transpose().reset_index(drop=True)
df_away = pd.concat([away_team[['match_id', 'home_away', 'id', 'name']], away_stats], axis=1)

# Match Statistics
df_match_stats = pd.concat([df_home, df_away])
df_match_stats.columns = [col.lower().replace(' ','_') for col in df_match_stats.columns]
```
